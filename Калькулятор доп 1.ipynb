```python

import re
a = ['ноль','один','два','три','четыре','пять','шесть','семь','восемь','девять']
b = ['одиннадцать','двенадцать','тринадцать','четырнадцать','пятнадцать','шестнадцать','семнадцать','восемнадцать','девятнадцать']
c = ['десять','двадцать','тридцать','сорок','пятьдесят','шестьдесят','семьдесят','восемьдесят','девяносто']
d = ['плюс','минус','умножить','разделить','остатокотделения на']

# списки для вывода ответа
e = ['сто','двести','триста','четыреста','пятьсот','шестьсот','семьсот','восемьсот','девятьсот']
f = ['одна','две','три','четыре','пять','шесть','семь','восемь','девять']
m = ['десятая','десятых','сотая','сотых','тысячная','тысячных']

s = input('Введите пример: ')
l = re.split(' ', s)
if 'на' in l:
    l.pop(l.index('на'))
for i in range(len(l)):  
    if l[i] in d:
        p = i   # p - индекс операции в списке 
        
# Функция для определения операндов
def operand(ind1,ind2): 

    def number(h):
        if l[h] in m[:2]:   # если дробная часть числа - ДЕСЯТИЧНЫЕ
            n = (f.index(l[h-1])+1)/10
        elif l[h] in m[2:4]:   # если дробная часть числа - СОТЫЕ
            if l[h] == 'сотая':   # т.е. число с дробной частью 0.01
                n = 1/100
            elif l[h] == 'сотых':    
                if l[h-1] in f and l[h-2] in c:   # т.е. число типа 5.33
                    n = ((c.index(l[h-2])+1)*10 + f.index(l[h-1])+1)/100
                elif l[h-1] in b:   # т.е. число типа 5.13
                    n = (b.index(l[h-1])+11)/100  
                elif l[h-1] in f:   # т.е. число типа 5.03
                    n = (f.index(l[h-1])+1)/100
        elif l[h] in m[4:]:   # если дробная часть числа - ТЫСЯЧНЫЕ
            if l[h] == 'тысячная':   # т.е. число с дробной частью 0.001
                n = 1/1000
            elif l[h] == 'тысячных':
                if l[h-1] in f and l[h-2] in c and l[h-3] in e:   # т.е. число типа 5.333
                    n = ((e.index(l[h-3])+1)*100 + (c.index(l[h-2])+1)*10 + f.index(l[h-1])+1)/1000
                elif l[h-1] in f and l[h-2] in e:   # т.е. число типа 5.303
                    n = ((e.index(l[h-2])+1)*100 + f.index(l[h-1])+1)/1000
                elif l[h-1] in f and l[h-2] in c:   # т.е. число типа 5.033
                    n = ((c.index(l[h-2])+1)*10 + f.index(l[h-1])+1)/1000
                elif l[h-1] in b and l[h-2] in e:   # т.е. число типа 5.313
                    n = ((e.index(l[h-2])+1)*100 + (b.index(l[h-1])+11))/1000
                elif l[h-1] in b:   # т.е. число типа 5.013
                    n = (b.index(l[h-1])+11)/1000
                elif l[h-1] in f:   # т.е. число типа 5.003
                    n = (f.index(l[h-1])+1)/1000
        return n
        
        
        
    for i in range(10):  
        for j in range(9):
            
            if ind1 >= 0:   # разбираем ПЕРВОЕ число
                
                if l[ind2] in d or l[ind2] == 'и':   # если первое число односоставное   
                    if l[ind1] == a[i]:
                        n = i
                        if 'и' in l[:p]:   # если первое число односоставное и дробное 
                            n += number(p-1)
                    elif l[ind1] == b[j]:
                        n = (j+1) + 10
                        if 'и' in l[:p]:  
                            n += number(p-1)
                    elif l[ind1] == c[j]:
                        n = (j+1)*10 
                        if 'и' in l[:p]:    
                            n += number(p-1) 
                elif l[ind2] not in d and l[ind2] != 'и':   # если первое число двусоставное 
                    if l[ind2] == a[i] and l[ind1] == c[j]:
                        n = (j+1)*10 + i
                        if 'и' in l[:p]:   # если первое число односоставное и дробное
                            n += number(p-1)
                        
            if ind1 < 0: # разбираем ВТОРОЕ число
  
                if 'и' not in l[p+1:]:   # если второе число ЦЕЛОЕ
                    if l[ind2] in d:   # если второе число односоставное
                        if l[ind1] == a[i]:
                            n = i
                        elif l[ind1] == b[j]:
                            n = (j+1) + 10
                        elif l[ind1] == c[j]:
                            n = (j+1)*10                 
                    elif l[ind2] not in d:   # если второе число двусоставное
                        if l[ind1] == a[i] and l[ind2] == c[j]: 
                            n = (j+1)*10 + i 
                elif 'и' in l[p+1:]:   # если второе число ДРОБНОЕ
                    if len(l[p+1:l.index('и',p+1)]) == 1:   # если второе число односовтавное и дробное
                        if l[p+1] == a[i]:
                            n = i
                            n += number(-1)
                        elif l[p+1] == b[j]:
                            n = (j+1) + 10
                            n += number(-1)
                        elif l[p+1] == c[j]:
                            n = (j+1)*10     
                            n += number(-1)            
                    elif len(l[p+1:l.index('и',p+1)]) == 2:   # если второе число двусоставное и дробное
                        if l[p+2] == a[i] and l[p+1] == c[j]: 
                            n = (j+1)*10 + i  
                            n += number(-1)
    return n
    
    
# Вывод результата
def calc(l):
    
    n1 = operand(0,1)     # цифровой вид первого число
    n2 = operand(-1,-2)   # цифровой вид второго числа
    
    # СЧИТАЕМ
    for i in range(len(d)):
        if l[p] == d[0]:
            rr = n1+n2
        elif l[p] == d[1]:
            rr = n1-n2
        elif l[p] == d[2]:
            rr = n1*n2
        elif l[p] == d[3]:
            rr = n1/n2
        else:
            r = n1%n2
    r = round(rr, 3)
    if r >= 0:                     # если ответ положительный 
        r1 = str(r)                
    else:                          # если ответ отрицательный 
        r1 = str(r).replace('-','')  
    rez = []      
    
    # Выводим результат в виде слов
    if '.' not in r1:   # число целое
        if len(r1) == 1:                
            rez.append(a[int(r1[-1])])
        if len(r1) >= 2:                
            if 10 < int(r1[-2:]) < 20:
                rez.append(b[int(r1[-1])-1])
            elif int(r1[-2:]) % 10 == 0:
                rez.append(c[int(r1[-2])-1])
            else:
                rez.append(a[int(r1[-1])])
                rez.append(c[int(r1[-2])-1])
        if len(r1) >= 3:
            rez.append(e[int(r1[-3])-1])
        if len(r1) >= 4:
            if int(r1[-4]) == 1:
                rez.append('тысяча')
            elif 1 <= int(r1[-4]) <= 4:
                rez.append('тысячи')
            else:
                rez.append('тысяч')
            rez.append(f[int(r1[-4])-1])
        if r < 0:
            rez.append('минус')
        rez.reverse()
    
    elif '.' in r1:   # число дробное
        if len(r1[r1.index('.')+1:]) == 1:
            if r1[-1] == '1':
                rez.append('десятая')
            else:
                rez.append('десятых')
        elif len(r1[r1.index('.')+1:]) == 2:
            if r1[-2:] == '01':
                rez.append('сотая')
            else:
                rez.append('сотых')
        elif len(r1[r1.index('.')+1:]) == 3:
            if r1[-3:] == '001':
                rez.append('тысячная')
            else:
                rez.append('тысячных')            
                
        if len(str(int(r1[r1.index('.')+1:]))) == 1:   # дробная часть числа 
            rez.append(f[int(r1[-1])-1])
        if len(str(int(r1[r1.index('.')+1:]))) >= 2:
            if int(r1[-1]) != 0 and int(r1[r1.index('.')+1:][-2]) != 1:            
                rez.append(f[int(r1[-1])-1])
            if int(r1[r1.index('.')+1:][-2]) != 0:                
                if int(r1[r1.index('.')+1:][-2]) == 1:
                    rez.append(b[int(r1[-1])-1])
                elif int(r1[r1.index('.')+1:][-2]) >= 2:
                    rez.append(c[int(r1[-2])-1])
        if len(str(int(r1[r1.index('.')+1:]))) >= 3:
            if int(r1[r1.index('.')+1:][-3]) != 0:
                rez.append(e[int(r1[-3])-1])
        if len(str(int(r1[r1.index('.')+1:]))) >= 4:
            if int(r1[-4]) == 1:
                rez.append('тысяча')
            elif 2 <= int(r1[-4]) <= 4:
                rez.append('тысячи') 
            elif 5 <= int(r1[-4]) <= 9:
                rez.append('тысяч')
            rez.append(f[int(r1[-4])-1])
        rez.append('и')
        
        if len(r1[:r1.index('.')]) == 1:   # целая часть числа               
            rez.append(a[int(r1[0])])
        if len(r1[:r1.index('.')]) >= 2:                
            if 10 < int(r1[:r1.index('.')]) < 20:
                rez.append(b[int(r1[r1.index('.')-1])-1])
            elif int(r1[:r1.index('.')]) % 10 == 0:
                rez.append(c[int(r1[r1.index('.')-2])-1])
            else:
                rez.append(a[int(r1[r1.index('.')-1])])
                rez.append(c[int(r1[r1.index('.')-2])-1])
        if len(r1[:r1.index('.')]) >= 3:
            rez.append(e[int(r1[r1.index('.')-3])-1])
        if len(r1[:r1.index('.')]) >= 4:
            if int(r1[r1.index('.')-4]) == 1:
                rez.append('тысяча')
            elif 1 <= int(r1[r1.index('.')-4]) <= 4:
                rez.append('тысячи')
            else:
                rez.append('тысяч')
            rez.append(f[int(r1[r1.index('.')-4])-1])
         
        if r < 0:
            rez.append('минус')
        rez.reverse()   
            
    print(*rez)    
calc(l) 
```
